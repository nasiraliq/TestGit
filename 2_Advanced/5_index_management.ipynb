# Index Management

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("../data/data_jobs.csv")
```

```python
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
```

```python
df.index.name = 'job_index'
```

```python
df.index.name
```

# Three Methods to cover
- `df.reset_index()`
- `df.set_index()`
- `df.sort_index()`

```python
df_pak = df[df["job_country"] == "Pakistan"].reset_index()
df_pak.head(3)
```

```python
df_pak.set_index(keys=["job_index"], inplace=True)
```
